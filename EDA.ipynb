```python
# Import core libraries for data handling and plotting
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the CSV and parse date columns
df = pd.read_excel("D:/Analytics/Practice datasets/Orders.xlsx", parse_dates=["Order Date", "Ship Date"])
```

```python
# Preview a few rows to understand the structure
df.head()
```

```python
# Check column types and non-null counts
df.info()
```

```python
# Get basic descriptive statistics for all columns
df.describe(include="all")
```

```python
# Check missing values in each column
df.isna().sum()
```

```python
# Check for duplicate rows based on Row ID (should be unique)
df.duplicated(subset=["Row ID"]).sum()
```

```python
# Convert relevant columns to categorical type for efficiency and clarity
cat_cols = [
    "Segment", "Category", "Sub-Category", "Region",
    "Country/Region", "City", "Ship Mode"
]
for c in cat_cols:
    df[c] = df[c].astype("category")
```

```python
# Summarize numeric fields to spot outliers and scale
num_cols = ["Sales", "Quantity", "Revenue", "Discount", "Profit"]
df[num_cols].describe()
```

```python
# Visualize distributions and outliers for numeric columns
sns.boxplot(data=df[num_cols])
plt.title("Numeric Columns – Boxplot")
plt.show()
```

```python
# Top 10 customers by total revenue
top_customers = (
    df.groupby("Customer ID")["Revenue"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
top_customers
```

```python
# Top 10 products by total revenue
top_products = (
    df.groupby("Product Name")["Revenue"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
top_products
```

```python
# Create additional time features from Order Date
df["Order_Year"] = df["Order Date"].dt.year
df["Order_Month"] = df["Order Date"].dt.to_period("M")
```

```python
# See dtype and a few sample values
print(df["Revenue"].dtype)
print(df["Revenue"].head(10))
```

```python
# Aggregate revenue and profit by month for trend analysis
rev_month = (
    df.groupby("Order_Month")[["Revenue", "Profit"]]
    .sum()
    .reset_index()
)
```

```python
#check dtypes of the DataFrame rev_month
print(rev_month.dtypes)
print(rev_month.head())
```

```python
# Convert Period[M] to timestamp for plotting
rev_month["Order_Month_dt"] = rev_month["Order_Month"].dt.to_timestamp()
```

```python
# Plot monthly revenue trend
sns.lineplot(data=rev_month.sort_values("Order_Month_dt"),
             x="Order_Month_dt", y="Revenue")
plt.title("Monthly Revenue Trend")
plt.xticks(rotation=45)
plt.show()

# Plot monthly profit trend
sns.lineplot(data=rev_month.sort_values("Order_Month_dt"),
             x="Order_Month_dt", y="Profit")
plt.title("Monthly Profit Trend")
plt.xticks(rotation=45)
plt.show()
```

```python
# Compute shipping delay in days
df["Ship_Days"] = (df["Ship Date"] - df["Order Date"]).dt.days

# Visualize distribution of shipping days
sns.histplot(df["Ship_Days"], kde=True)
plt.title("Distribution of Shipping Days")
plt.show()
```

```python
# Aggregate revenue and profit by key business dimensions
group_cols = ["Category", "Sub-Category", "Region", "Segment", "Ship Mode"]

for c in group_cols:
    print(f"\n=== Revenue and Profit by {c} ===")
    tmp = (
        df.groupby(c, observed=False)[["Revenue", "Profit"]]
        .sum()
        .sort_values("Profit", ascending=False)
    )
    print(tmp.head(10))
```

```python
# Top 10 products and customers by profit
top_profit_products = (
    df.groupby("Product Name")["Profit"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
top_profit_customers = (
    df.groupby("Customer ID")["Profit"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

top_profit_products, top_profit_customers
```

```python
# Calculate profit margin as Profit / Revenue
df["Profit_Margin"] = df["Profit"] / df["Revenue"]

# Visualize profit margin distribution
sns.histplot(df["Profit_Margin"], kde=True)
plt.title("Distribution of Profit Margin")
plt.show()
```

```python
def get_top_cities_by_revenue(dataframe: pd.DataFrame, n: int = 10) -> pd.DataFrame:
    """
    Compute the top N cities by total revenue.
    
    Args:
        dataframe (pd.DataFrame): Input DataFrame containing 'City' and 'Revenue' columns.
        n (int, optional): Number of top cities to return. Defaults to 10.
        
    Returns:
        pd.DataFrame: DataFrame containing top N cities sorted by revenue in descending order.
    """
    top_cities = (
        dataframe.groupby("City", observed=False, as_index=False)["Revenue"]
        .sum()
        .sort_values("Revenue", ascending=False)
        .head(n)
    )
    return top_cities


def plot_top_cities_by_revenue(top_cities: pd.DataFrame) -> None:
    """
    Create a bar chart of top cities by revenue.
    
    Args:
        top_cities (pd.DataFrame): DataFrame containing 'City' and 'Revenue' columns.
    """
    import matplotlib.pyplot as plt
    
    plt.figure(figsize=(12, 6))
    plt.bar(range(len(top_cities)), top_cities['Revenue'], color='skyblue')
    plt.title('Top 10 Cities by Revenue')
    plt.xlabel('City Rank')
    plt.ylabel('Revenue')
    plt.xticks(range(len(top_cities)), top_cities['City'], rotation=45, ha='right')
    plt.tight_layout()
    plt.show()


# Get and display top 10 cities by revenue
top_cities = get_top_cities_by_revenue(df)
print("top_cities shape:", top_cities.shape)
print(top_cities)

# Visualize the results
plot_top_cities_by_revenue(top_cities)
```
